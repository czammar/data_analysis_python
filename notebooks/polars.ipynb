**Conteo de vuelos por fecha**

* Obtenemos las columnas de fecha y estado
* creamos una pivot para contar
* obtenemos los 30 primeros registros

```python
%%time 
data.\
    select(pl.col("FlightDate", "OriginStateName")).\
    group_by("FlightDate", "OriginStateName").\
    len().\
    sort(by="len").\
    top_k(k=30, by="len", reverse=False)
```

Ahora escribirmos al resultado a `output.csv`

```python
%%time 
data.\
    select(pl.col("FlightDate", "OriginStateName")).\
    group_by("FlightDate", "OriginStateName").\
    len().\
    sort(by="len").\
    top_k(k=30, by="len", reverse=False).\
    write_csv("output.csv")
```
